```python
import pulp
import csv

# Crear un problema de programación lineal
problem = pulp.LpProblem("Optimizacion de Horarios", pulp.LpMinimize)

# Definir las franjas horarias (de 8:00 am a 6:00 pm)
horas = list(range(8, 18))

# Duración mínima y máxima de una franja de trabajo
duracion_minima = 1
duracion_maxima = 2

# Crear variables binarias para representar el estado de cada empleado en cada franja horaria
estados = ["Trabaja", "Pausa Activa", "Almuerza", "Nada"]
empleados_estado = {}
for empleado in range(1, 9):
    for hora in horas:
        for estado in estados:
            empleados_estado[(empleado, estado, hora)] = pulp.LpVariable(f"Empleado_{empleado}_{estado}_Hora_{hora}", cat=pulp.LpBinary)

# Función objetivo: minimizar el tiempo en estado Nada
problem += pulp.lpSum(empleados_estado[(empleado, "Nada", hora)] for empleado in range(1, 9) for hora in horas), "Minimizar el tiempo en estado Nada"

# Restricciones
for empleado in range(1, 9):
    for hora in horas:
        # Cada empleado está en un estado en cada franja horaria
        problem += pulp.lpSum(empleados_estado[(empleado, estado, hora)] for estado in estados) == 1

        # Duración mínima y máxima de una franja de trabajo
        franja_trabajo = pulp.lpSum(empleados_estado[(empleado, "Trabaja", hora)] for hora in horas)
        problem += duracion_minima <= franja_trabajo <= duracion_maxima

        # Restricciones de almuerzo
        if 11 <= hora <= 14:  # De 11:00 am a 2:00 pm
            almuerza = pulp.lpSum(empleados_estado[(empleado, "Almuerza", hora)] for hora in range(11, 14))
            problem += almuerza >= 1  # Debe tomar almuerzo
            problem += almuerza <= 1  # Debe tomar almuerzo una única vez
            if hora < 13:  # De 11:00 am a 12:59 pm
                problem += almuerza >= 1  # Hora mínima para tomar almuerzo
            if hora > 13:  # De 1:00 pm a 2:00 pm
                problem += almuerza <= 1  # Hora máxima para tomar almuerzo

# Restricciones de transición de estados
for empleado in range(1, 9):
    for hora in horas:
        if hora < 17:  # No aplicar restricciones en la última franja horaria
            for estado in estados:
                if estado == "Nada":
                    problem += empleados_estado[(empleado, estado, hora + 1)] - empleados_estado[(empleado, estado, hora)] >= 0
                else:
                    problem += empleados_estado[(empleado, estado, hora + 1)] - empleados_estado[(empleado, estado, hora)] <= 0

# Último estado de la jornada laboral debe ser Trabaja
for empleado in range(1, 9):
    problem += empleados_estado[(empleado, "Trabaja", 17)] == 1

# Debe haber al menos 1 empleado trabajando en cada franja horaria
for hora in horas:
    problem += pulp.lpSum(empleados_estado[(empleado, "Trabaja", hora)] for empleado in range(1, 9)) >= 1

# Resolver el problema
problem.solve()

# Imprimir los resultados
for empleado in range(1, 9):
    for hora in horas:
        for estado in estados:
            if pulp.value(empleados_estado[(empleado, estado, hora)]) == 1:
                print(f"Empleado {empleado} {estado} en la franja horaria {hora}:00 - {hora + 1}:00")
    
with open("resultados.csv", "w", newline="") as csvfile:
    fieldnames = ["numero_empleado", "hora", "estado"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for empleado in range(1, 9):
        for hora in horas:
            for estado in estados:
                if pulp.value(empleados_estado[(empleado, estado, hora)]) == 1:
                    writer.writerow({"numero_empleado": empleado, "hora": hora, "estado": estado})

# Resolver el problema
# problem.solve()

# Guardar los resultados en un archivo CSV


print("Resultados guardados en resultados.csv")

```
